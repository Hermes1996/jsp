```python
#!/usr/bin/python3
# -*- coding: utf-8 -*-
import requests
import json

# 豆瓣最热电影，network下面
url = "https://movie.douban.com/j/search_subjects"
# 定义请求头
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36"
}
# 循环构建请求参数并且发送请求
for page_start in range(0, 20, 20):
    params = {
        "type": "tv",
        "tag": "热门",
        "sort": "recommend",
        "page_limit": "20",
        "page_start": page_start
    }
    response = requests.get(
        url=url,
        headers=headers,
        params=params
    )
    # 方式一:直接转换json方法
    # results = response.json()
    # 方式二: 手动转换
    # 获取字节串
    content = response.content
    # 转换成字符串
    string = content.decode('utf-8')
    # 把字符串转成python数据类型
    results = json.loads(string)
    # 解析结果,评分逆序
    #     {"subjects":[{"episodes_info":"36集全","rate":"8.3","cover_x":1080,"title":"御赐小仵作","url":"https:\/\/movie.douban.com\/subject\/32581207\/","playable":true,"cover":"https://img3.doubanio.com\/view\/photo\/s_ratio_poster\/public\/p2642463440.webp","id":"32581207","cover_y":1656,"is_new":false},
    for movie in sorted(results["subjects"],key=lambda keys:keys['rate'], reverse=True):
        print(movie["title"], movie["rate"])
```
